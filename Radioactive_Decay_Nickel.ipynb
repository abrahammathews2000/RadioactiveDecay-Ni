```python
# Importing necessary libraries
import numpy as np
import scipy
from scipy import constants
from scipy import special
from scipy.integrate import solve_ivp
import math
import matplotlib.pyplot as plt

plt.style.use('default')
```

```python
tNi = 6.1  #days
tCo = 77.7 #days

def odeEqn(t,X):
    x,y,z = X
    f1 = -(x/tNi) 
    f2 = (x/tNi)-(y/tCo)
    f3 = (y/tCo)
    return[f1,f2,f3]

x_0 = 100
y_0 = 0
z_0 = 0
X_0 = [x_0,y_0,z_0]

soln = solve_ivp(odeEqn,(0,300),X_0,max_step=(1/500))
```

```python
#print(soln.t)
#print(soln.y)
fig = plt.figure(figsize = (12,7))

plt.plot(soln.t,soln.y[0],label = "Nickel",linestyle = "solid")
plt.plot(soln.t,soln.y[1],label = "Cobalt",linestyle = "dashed")
plt.plot(soln.t,soln.y[2],label = "Iron",linestyle = "dashdot")

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title("Formation of Fe from the decay of Nickel and Cobalt in Supernova")
plt.xlabel("Time (days)")
plt.ylabel("N")
plt.grid()
plt.show()
```
